```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(palette="Spectral")
```

```python
df = pd.read_csv("../stats.csv")
```

```python
df = df.round(3)
```

```python
df
```

```python
df.RMSE = df.RMSE * (-1)
```

```python
df
```

```python
sns.set(palette="Spectral")
g = sns.relplot(data = df, x="Algorithm",y="CPU", col="Ordinal",col_wrap=2, hue="Log", kind="line", height=6, )
g.set_xticklabels(rotation=30)
```

```python
sns.set(palette="Spectral")
g = sns.catplot(data = df, x="Algorithm",y="CPU", col="Ordinal",col_wrap=2, hue="Log", kind="bar", height=6, ci=False)
g.set_xticklabels(rotation=90)
```

```python
sns.set(palette="Pastel1")
g = sns.relplot(data = df, x="Algorithm",y="RMSE", col="Ordinal",col_wrap=2, hue="Log", kind="line", height=6 )
g.set_xticklabels(rotation=30)
plt.ylim(0.08,0.165)
```

```python
sns.set(palette="Pastel1")
g = sns.relplot(data = df, x="Algorithm",y="RMSE",hue="Ordinal", palette=["blue", "darkblue"], kind="line", height=6,ci = False )
g.set_xticklabels(rotation=90)
plt.ylim(0.08,0.165)
```

```python
sns.set(palette="Pastel1")
g = sns.catplot(data = df, x="Algorithm",y="RMSE", col="Ordinal",col_wrap=1, hue="Log", kind="bar", height=6 )
g.set_xticklabels(rotation=30)
plt.ylim(0.08,0.165)
```

```python
sns.set(palette="Accent")
g = sns.relplot(data = df, x="Algorithm",y="R2", col="Ordinal",col_wrap=2, hue="Log", kind="line", height=6 )
g.set_xticklabels(rotation=90)
plt.ylim(0.8,0.95)
```

```python
sns.set(palette="Accent")
g = sns.catplot(data = df, x="Algorithm",y="R2", col="Ordinal",col_wrap=2, hue="Log", kind="bar", height=6,  ci=False)
g.set_xticklabels(rotation=30)
plt.ylim(0.8,0.95)
```

```python
sns.set(palette="Pastel1")
g = sns.relplot(data = df, x="Algorithm",y="R2",hue="Ordinal", palette=["lightgreen", "green"], kind="line", height=6,ci = False )
g.set_xticklabels(rotation=90)
plt.ylim(0.8,0.95)
```
